VISUALIZE DATA USING APPROPRIATE LIBRARIES

- CREATING BASIC VISUALIZATIONS WITH MATPLOTLIB

```python
Goals = footBall['Goals'].loc[100:110]
position = [100, 101, 102, 103, 104, 105, 106, 107, 108, 109, 110]
```

```python
plt.figure(figsize = (10, 5))
plt.plot(position, Goals, marker='*', linestyle='-', color='green')
plt.title('The Goals for players with index positions 100 to 110.')
plt.xlabel("The index positions 100 to 110")
plt.ylabel("Goals")
plt.grid(True)
plt.show()
```

CREATING A SCATTER PLOT

```python
JerseyNumber = footBall['Jersey Number'].loc[:9]
playerName = footBall['Name'].loc[:9]
```

```python
plt.figure(figsize=(15,5))
plt.scatter(playerName, JerseyNumber, marker='*', color='red')
plt.title('Player Name and Jersey Number')
plt.xlabel('Jersey Number')
plt.ylabel('Player Name')
plt.grid(True)
plt.show()
```

Histogram

```python
data = np.random.normal(loc=20.0, scale=5.0, size=365)

plt.hist(data, bins=20, color='green', alpha=0.7)
plt.title('Histogram of Daily Temperatures Over One Year')
plt.xlabel('Temperature (degrees Celsius)')
plt.ylabel('Frequency')
plt.show()
```

CREATING A BAR CHART

```python
Name = footBall['Name'].loc[:4]
Age = footBall['Age'].loc[:4]
```

```python
plt.figure(figsize=(8, 5))
plt.bar(Name, Age, color='red', alpha=0.7)
plt.title("Foot Ball player's Name and Age")
plt.xlabel('Names')
plt.ylabel('Age')
plt.show()
```

CREATING A PIE CHART

```python
Wins = footBall['Wins'].loc[:4]
Nation = footBall['Nationality'].loc[4]
```

```python
plt.figure(figsize=(7,7))
plt.pie(Wins, labels=Nation, autopct='%1.1f%%', startangle=140)
plt.title('Smartphone Market Share')
plt.show()
```

BOXPLOT: ANALYSE THE DISTRIBUTION OF TEST SCORE

```python
scores = np.random.normal(7, 1, 10)
scores = np.append(scores, footBall[footBall['Wins'] == 'Arsenal'].loc[:4])
```

```python
plt.figure(figsize=(6,8))
plt.boxplot(scores, vert=True, patch_artist=True, meanline=True, showmeans=True)
plt.title('Distribution Wins for Arsenal')
plt.ylabel('Match Won')
plt.grid(True)
plt.show()
```
